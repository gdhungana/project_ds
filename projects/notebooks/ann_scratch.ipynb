```python
from matplotlib import pyplot as plt
import numpy as np
from projects import ann_from_scratch as ann
```

```python
#- plot an example activation function -- Sigmoid fn.
x = np.arange(-10, 10, 0.1)
f=ann.activation_fn(x)
plt.plot(x, f)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
# function with weights
w1 = 0.5
w2 = 1.0
w3 = 2.0
l1 = 'w = 0.5'
l2 = 'w = 1.0'
l3 = 'w = 2.0'
for w, l in [(w1, l1), (w2, l2), (w3, l3)]:
    f = ann.weighted_biased_act_fn(x,w=w)
    plt.plot(x, f, label=l)
plt.xlabel('x')
plt.ylabel('h_w(x)')
plt.legend(loc=2)
plt.show()
```

```python
#Function with bias
b1 = -2.0
b2 = 0
b3 = 2.0
l1 = 'b = -2.0'
l2 = 'b = 1.0'
l3 = 'b = 2.0'
for b, l in [(b1, l1), (b2, l2), (b3, l3)]:
    f = ann.weighted_biased_act_fn(x,w=1,b=b)
    plt.plot(x, f, label=l)
plt.xlabel('x')
plt.ylabel('h_b(x)')
plt.legend(loc=2)
plt.show()
```

```python
# - set up a network with 3 layers
#- first layer with three nodes
#- weight is a matrix, bias is a vector. dimensionality comes from the nodes 

w1 = np.array([[0.2, 0.2, 0.2], [0.4, 0.4, 0.4], [0.6, 0.6, 0.6]])
b1 = np.array([0.8, 0.8, 0.8])
w2 = np.array([[0.5, 0.5, 0.5]])
b2 = np.array([0.2])

#- Organize the full list of weights and biases
w = [w1, w2]
b = [b1, b2]
#- input x vector:
x = [1.5, 2.0, 3.0]
n_layers=3
h_3=ann.looped_nn_calc(n_layers, x, w, b)

print(h_3)
```
